```python
from collections import defaultdict
all_nouns = Counter()
noun_dict = {}
noun_dict_list = defaultdict(set)
for text in b_reviews.text:
    doc = nlp(text)
    for chunk in doc.noun_chunks:
        root_text = chunk.root.text.strip().lower().replace("\n", " ")
        all_text = chunk.text.strip().lower().replace("\n", " ")
        if root_text not in noun_dict or all_text.count(' ') > noun_dict[root_text].count(' '):
            noun_dict[root_text] = all_text
        noun_dict_list[root_text].add(all_text)
        all_nouns.update([root_text])
```

```python
from wordfreq import word_frequency

new_noun_scores = {}
for noun, freq in all_nouns.items():
    word_freq = word_frequency(noun_dict[noun], 'en')
    score = freq / word_freq if word_freq != 0 else freq
    new_noun_scores[noun] = score
```

```python
list(enumerate(Counter(new_noun_scores).most_common()))
```

```python
[(i, x[0], noun_dict_list[x[0]]) for i, x in enumerate(Counter(new_noun_scores).most_common())]
```
